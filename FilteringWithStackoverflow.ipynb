## To filter salary having greate than 70,000 USD (Converted Comp Is salary Col)

> we have make a filt variable and used our condition to get a boolean df then passing it in loc

```python
filt = (df['ConvertedComp'] > 70000)
```

> in Loc we can pass 2nd arg as col so <code>df.loc[row , col] </code>

```python
df.loc[filt , ['Country' , 'LanguageWorkedWith' , 'ConvertedComp']]
```
